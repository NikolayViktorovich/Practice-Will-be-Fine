# Тема: "Набор данных фишинга для машинного обучения"

## Описание проекта
Фишинг - это один из распространённых видов кибератак, при котором злоумышленники создают поддельные сайты, имитирующие, например, сайты банков, соцсетей или платежные системы в фальшивых интернет-магазинах, чтобы похитить данные юзеров: логины, пароли, реквизиты банковских карт и т.д. Обнаружение фишинга - важная задача в области кибербезопасности, поскольку такие атаки наносят значительный ущерб как частным лицам, так и организациям.

# Цель и задачи проекта
**Цель проекта:**
Разработать модель машинного обучения для автоматической классификации сайтов на фишинговые и легитимные на основе характеристик URL, домена и HTML структуры страницы.

**Задачи проекта:**
- загрузка и ознакомление с данными;
- предварительная обработка;
- полноценный разведочный анализ;
- создание эффективной модели классификации для обнаружения фишинговых сайтов;
- обеспечение возможности интеграции модели в системы кибербезопасности

# Типы данных и описание полей данных

- **id**: `int64` — Уникальный идентификатор записи в датасете.
- **NumDots**: `int64` — Количество точек (`.`) в URL (например, в `sub.domain.com` — 3 точки).
- **SubdomainLevel**: `int64` — Уровень поддоменов в URL (например, `sub.sub.domain.com` — 2 поддомена).
- **PathLevel**: `int64` — Количество уровней в пути URL (например, `/path1/path2/path3` — 3 уровня).
- **UrlLength**: `int64` — Общая длина URL в символах.
- **NumDash**: `int64` — Количество дефисов (`-`) в URL.
- **NumDashInHostname**: `int64` — Количество дефисов (`-`) в имени хоста URL.
- **AtSymbol**: `int64` — Наличие символа `@` в URL (0 — отсутствует, 1 — присутствует).
- **TildeSymbol**: `int64` — Наличие символа `~` в URL (0 — отсутствует, 1 — присутствует).
- **NumUnderscore**: `int64` — Количество символов подчеркивания (`_`) в URL.
- **NumPercent**: `int64` — Количество символов процента (`%`) в URL.
- **NumQueryComponents**: `int64` — Количество компонентов в строке запроса URL (например, `key1=value1&key2=value2` — 2 компонента).
- **NumAmpersand**: `int64` — Количество символов амперсанда (`&`) в URL.
- **NumHash**: `int64` — Количество символов решетки (`#`) в URL.
- **NumNumericChars**: `int64` — Количество числовых символов в URL.
- **NoHttps**: `int64` — Отсутствие HTTPS в URL (0 — HTTPS присутствует, 1 — отсутствует).
- **RandomString**: `int64` — Наличие случайной строки в URL (0 — отсутствует, 1 — присутствует).
- **IpAddress**: `int64` — Наличие IP-адреса вместо доменного имени в URL (0 — домен, 1 — IP).
- **DomainInSubdomains**: `int64` — Наличие основного домена в поддоменах (0 — отсутствует, 1 — присутствует).
- **DomainInPaths**: `int64` — Наличие основного домена в пути URL (0 — отсутствует, 1 — присутствует).
- **HttpsInHostname**: `int64` — Наличие `https` в имени хоста URL (0 — отсутствует, 1 — присутствует).
- **HostnameLength**: `int64` — Длина имени хоста в URL (в символах).
- **PathLength**: `int64` — Длина пути в URL (в символах).
- **QueryLength**: `int64` — Длина строки запроса в URL (в символах).
- **DoubleSlashInPath**: `int64` — Наличие двойного слэша (`//`) в пути URL (0 — отсутствует, 1 — присутствует).
- **NumSensitiveWords**: `int64` — Количество "чувствительных" слов в URL (например, `login`, `password`).
- **EmbeddedBrandName**: `int64` — Наличие имени известного бренда в URL (0 — отсутствует, 1 — присутствует).
- **PctExtHyperlinks**: `float64` — Доля внешних гиперссылок на веб-странице (от 0.0 до 1.0).
- **PctExtResourceUrls**: `float64` — Доля внешних URL-адресов ресурсов (например, изображений, скриптов) на странице.
- **ExtFavicon**: `int64` — Использование внешнего favicon (0 — внутренний, 1 — внешний).
- **InsecureForms**: `int64` — Наличие небезопасных форм (например, отправка данных по HTTP) (0 — отсутствуют, 1 — присутствуют).
- **RelativeFormAction**: `int64` — Использование относительных URL в действиях форм (0 — абсолютные, 1 — относительные).
- **ExtFormAction**: `int64` — Использование внешних URL в действиях форм (0 — внутренние, 1 — внешние).
- **AbnormalFormAction**: `int64` — Наличие аномальных действий форм (например, пустые или некорректные URL) (0 — нормальные, 1 — аномальные).
- **PctNullSelfRedirectHyperlinks**: `float64` — Доля гиперссылок, ведущих на `null` или саму страницу.
- **FrequentDomainNameMismatch**: `int64` — Частое несоответствие доменных имен в ссылках и основном домене (0 — нет, 1 — есть).
- **FakeLinkInStatusBar**: `int64` — Наличие поддельных ссылок в строке состояния (0 — отсутствуют, 1 — присутствуют).
- **RightClickDisabled**: `int64` — Отключение правого клика на странице (0 — разрешено, 1 — отключено).
- **PopUpWindow**: `int64` — Наличие всплывающих окон (0 — отсутствуют, 1 — присутствуют).
- **SubmitInfoToEmail**: `int64` — Отправка данных формы на email (0 — нет, 1 — да).
- **IframeOrFrame**: `int64` — Наличие iframe или frame на странице (0 — отсутствуют, 1 — присутствуют).
- **MissingTitle**: `int64` — Отсутствие заголовка страницы (0 — есть, 1 — отсутствует).
- **ImagesOnlyInForm**: `int64` — Формы содержат только изображения (0 — нет, 1 — да).
- **SubdomainLevelRT**: `int64` — Результат теста на уровень поддоменов (например, -1, 0, 1; возможно, категориальный индикатор риска).
- **UrlLengthRT**: `int64` — Результат теста на длину URL (например, -1, 0, 1; индикатор риска).
- **PctExtResourceUrlsRT**: `int64` — Результат теста на долю внешних ресурсов (например, -1, 0, 1; индикатор риска).
- **AbnormalExtFormActionR**: `int64` — Результат теста на аномальные действия форм (например, -1, 0, 1; индикатор риска).
- **ExtMetaScriptLinkRT**: `int64` — Результат теста на внешние мета-теги, скрипты или ссылки (например, -1, 0, 1; индикатор риска).
- **PctExtNullSelfRedirectHyperlinksRT**: `int64` — Результат теста на долю `null` или самонаправленных гиперссылок (например, -1, 0, 1; индикатор риска).
- **CLASS_LABEL**: `int64` — Метка класса (0 — легитимный сайт, 1 — фишинговый сайт).

## Выявление проблем в данных
### 1.4 Беглый осмотр данных для выявления проблем

### Проблемы в данных:
- Наименования столбцов используют CamelCase, а не snake_case, ЭТО НЕ СООТВЕТСТВУЕТ стандартам python;
- Возможны пропуски в данных (будут обязательно проверены далее);
- Типы данных могут быть некорректными;
- Нужно провести проверки на дубликаты;
- Некоторые признаки как пример, PctExtHyperlinks, PctExtResourceUrls имеют значения в диапазоне (0, 1), что указывает на нормализацию НО нужно проверить их правильность.
- Столбец CLASS_LABEL имеет значения (-1, 0, 1), где 0 (подозрительный) может быть неоднозначным для классификации

**Вывод:** были обнаружены след. проблемы: названия столбцов в CamelCase, потенциальные пропуски, возможные дубликаты, неоднозначность значений в class_label (-1, 0, 1). Все остальные аспекты будут проверены далее.

## Коррекция наименований столбцов
### 1.5 Приведение названия столбцов к snake_case

```python
df.columns = (df.columns
              .str.replace(r'(?<!^)(?=[A-Z])', '_', regex=True)
              .str.lower())
print("\nНаименования после приведения к snake_case:")
print(df.columns.tolist())
```

### 1.6 Приведение текстовых данных к нижнему регистру

```python
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.lower()
print("\nПриведено к нижнему регистру (если такие данные вообще были)")
```

**Вывод:** названия столбцов были приведены к snake_case и нижнему регистру, текстовых столбцов нет.

## Проверка и обработка пропусков
### 1.7 Проверка пропусков

```python
print("\nПропуски:")
display(df.isnull().sum())
```

### 1.8 Процент пропусков

```python
missing_percentage = df.isnull().sum() / len(df) * 100
print("\nПроцент пропусков по столбцам:")
display(missing_percentage)
```

### 1.9 Обработка пропусков

```python
if missing_percentage.max() < 5:
    df = df.dropna()
    print("\nСтроки с пропусками удалены, т.к они составляют менее 5% данных или отсутствуют")
else:
    for col in df.select_dtypes(include=['float64', 'int64']).columns:
        df[col] = df[col].fillna(df[col].median())
    print("\nПропуски в числовых столбцах заполнены медианами")
```

### 1.10 Проверка после обработки пропусков

```python
print("\nПропуски после обработки:")
display(df.isnull().sum())
```

**Вывод:** после проверки все атрибуты имеют нулевое количество пропусков после обработки. Пропуски отсутствуют. Нулевые значения не требуют замены, т.к они информативны

### 1.12 Изменение типов данных

```python
binary_cols = ['at_symbol', 'tilde_symbol', 'no_https', 'random_string', 'ip_address',
               'domain_in_subdomains', 'domain_in_paths', 'https_in_hostname',
               'ext_favicon', 'insecure_forms', 'relative_form_action', 'ext_form_action',
               'abnormal_form_action', 'fake_link_in_status_bar', 'right_click_disabled',
               'pop_up_window', 'submit_info_to_email', 'iframe_or_frame', 'missing_title',
               'images_only_in_form', 'subdomain_level_rt', 'url_length_rt',
               'abnormal_ext_form_action_r', 'class_label']
for col in binary_cols:
    if col in df.columns:
        df[col] = df[col].astype('int8')

float_cols = ['pct_ext_hyperlinks', 'pct_ext_resource_urls', 'pct_null_self_redirect_hyperlinks',
              'pct_ext_null_self_redirect_hyperlinks_rt']
for col in float_cols:
    if col in df.columns:
        df[col] = df[col].astype('float32')
```

### 3.1 Графический анализ с Matplotlib

```python
# Диаграмма 1 (гистограмма длины URL для каждого класса)
plt.hist(df[df['c_l_a_s_s__l_a_b_e_l'] == 0]['url_length'], bins=30, alpha=0.5, label='Легитимные', color='green')
plt.hist(df[df['c_l_a_s_s__l_a_b_e_l'] == 1]['url_length'], bins=30, alpha=0.5, label='Фишинговые', color='red')
plt.title('Распределение длины URL для фишинговых и легитимных сайтов', fontsize=14, pad=10)
plt.xlabel('Длина URL', fontsize=12)
plt.ylabel('Количество сайтов', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.savefig('url_length_histogram.png')
plt.show()
```

**Вывод:** Фишинговые сайты имеют более длинные URL (НАИБОЛЕЕ ЧАСТО больше 100 символов), тогда как легитимные короче (50–70 символов) что подтверждает длину URL как ключевой признак фишинга.

```python
# Диаграмма 2 (столбчатая диаграмма среднего количества точек)
mean_dots = df.groupby('c_l_a_s_s__l_a_b_e_l')['num_dots'].mean()
plt.bar(['Легитимные', 'Фишинговые'], mean_dots, color=['green', 'red'])
plt.title('Среднее количество точек в URL по классам', fontsize=14, pad=10)
plt.ylabel('Среднее количество точек', fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.savefig('num_dots_bar.png')
plt.show()
```

**Вывод:** Наиболее хитренькие фишинговые сайты имеют больше точек в URL (3-4 против 1–2 у легитимных) что естественно указывает на сложные доменные структуры для обмана.

```python
# Диаграмма 3 (круговая диаграмма для no_https)
https_counts = df.groupby(['c_l_a_s_s__l_a_b_e_l', 'no_https']).size().unstack()
labels = ['Легитимные (HTTPS)', 'Легитимные (без HTTPS)', 'Фишинговые (HTTPS)', 'Фишинговые (без HTTPS)']
sizes = [https_counts.loc[0, 0], https_counts.loc[0, 1], https_counts.loc[1, 0], https_counts.loc[1, 1]]
colors = ['green', 'lightgreen', 'red', 'pink']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 10})
plt.title('Доля сайтов с/без HTTPS по классам', fontsize=14, pad=10)
plt.axis('equal')
plt.tight_layout()
plt.savefig('https_pie.png')
plt.show()
```

**Вывод:** Фишинги чаще обходят HTTPS (высокая доля no_https=1) в отличие от легитимных, что особенно подчеркивает их просто ужасную ненадежность.
